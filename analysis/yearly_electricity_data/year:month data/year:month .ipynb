```python
# 1. Define the variables we need for the Energy Security task
security_vars = ['Demand', 'Total Generation']

# 2. Filter the Global and Europe DataFrames
# This makes the data much smaller and faster to work with
df_global_security = df_global_yearly[df_global_yearly['Variable'].isin(security_vars)]
df_europe_security = df_europe_yearly[df_europe_yearly['Variable'].isin(security_vars)]

# 3. Quick check to see the first few rows
print(f"Global Security Data Rows: {len(df_global_security)}")
display(df_global_security.head())
```

```python
# Create the pivoted 'Wide' DataFrame
df_pivot = df_global_security.pivot_table(
    index=['Area', 'ISO 3 code', 'Year', 'Continent'], 
    columns='Variable', 
    values='Value'
).reset_index()

# Rename columns for easier math
df_pivot.columns.name = None  # Remove the hierarchy name
df_pivot = df_pivot.rename(columns={
    'Demand': 'demand_twh',
    'Total Generation': 'generation_twh'
})

# Preview the transformation
df_pivot.head()
```

```python
# 1. Calculate Energy Security Percentage
# (Generation / Demand) * 100
df_pivot['energy_security_pct'] = (df_pivot['generation_twh'] / df_pivot['demand_twh']) * 100

# 2. Basic Cleaning: Remove rows with NaN (where one of the two values was missing)
df_pivot = df_pivot.dropna(subset=['energy_security_pct'])

# 3. Quick EDA: Look at the most "secure" countries in the most recent year
latest_year = df_pivot['Year'].max()
print(f"--- Top 5 Energy Secure Countries in {latest_year} ---")
display(df_pivot[df_pivot['Year'] == latest_year].sort_values('energy_security_pct', ascending=False).head())
```
