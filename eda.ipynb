```python
# 3. Check for Missing Values
print("Missing Values:")
print(energy_access.isnull().sum())
print("\nMissing Values Percentage:")
print((energy_access.isnull().sum() / len(energy_access)) * 100)
```

```python
# 4. Visualize Missing Data
msno.matrix(energy_access, figsize=(12, 6))
plt.title("Missing Data Pattern")
plt.show()
```

```python
# Check missing values percentage by year
missing_pct = (energy_access.isnull().sum() / len(energy_access)) * 100
year_columns = [col for col in energy_access.columns if col.isdigit()]
year_missing = missing_pct[year_columns]

print("Years with data availability:")
print(year_missing[year_missing < 50].sort_values())
```

```python
# Check what indicators we have
print("Unique Indicators:")
print(energy_access['Indicator Name'].unique())
print(f"\nNumber of countries: {energy_access['Country Name'].nunique()}")
```

```python
# Focus on years with substantial data (likely 1990 onwards)
recent_years = [col for col in year_columns if int(col) >= 1990]
energy_recent = energy_access[['Country Name', 'Country Code', 'Indicator Name'] + recent_years]

print(f"Data from 1990 onwards:")
print(f"Shape: {energy_recent.shape}")
print(f"\nMissing values summary:")
print(energy_recent[recent_years].isnull().sum().describe())
```

```python
# Focus on the reliable data period (1990-2023)
reliable_years = [str(year) for year in range(1990, 2024)]
energy_clean = energy_access[['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'] + reliable_years].copy()

print(f"Clean dataset shape: {energy_clean.shape}")
print(f"\nMissing values in recent years:")
print(energy_clean[reliable_years].isnull().sum().tail(10))
```

```python
# Sample countries to see the data
print("Sample data:")
energy_clean.head(10)
```

```python
recent_years_subset = [str(year) for year in range(2010, 2024)]
print("Summary statistics (2010-2023):")
energy_clean[recent_years_subset].describe()
```

```python
# Check distribution of access rates in 2023
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
energy_clean['2023'].hist(bins=30, edgecolor='black')
plt.xlabel('Access to Electricity (%)')
plt.ylabel('Number of Countries')
plt.title('Distribution of Electricity Access in 2023')
```

```python
# Compare 1990 vs 2023
comparison = energy_clean[['Country Name', '1990', '2023']].dropna()
comparison['Change'] = comparison['2023'] - comparison['1990']
comparison = comparison.sort_values('Change', ascending=False)

print("Top 10 countries with biggest improvement:")
print(comparison.head(10))
```
