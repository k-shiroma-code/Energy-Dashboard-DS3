```python
# 1. Define the variables we need for the Energy Security task
security_vars = ['Demand', 'Total Generation']

# 2. Filter the Global and Europe DataFrames
# This makes the data much smaller and faster to work with
df_global_security = df_global_yearly[df_global_yearly['Variable'].isin(security_vars)]
df_europe_security = df_europe_yearly[df_europe_yearly['Variable'].isin(security_vars)]

# 3. Quick check to see the first few rows
print(f"Global Security Data Rows: {len(df_global_security)}")
display(df_global_security.head())
```

```python
# Create the pivoted 'Wide' DataFrame
df_pivot = df_global_security.pivot_table(
    index=['Area', 'ISO 3 code', 'Year', 'Continent'], 
    columns='Variable', 
    values='Value'
).reset_index()

# Rename columns for easier math
df_pivot.columns.name = None  # Remove the hierarchy name
df_pivot = df_pivot.rename(columns={
    'Demand': 'demand_twh',
    'Total Generation': 'generation_twh'
})

# Preview the transformation
df_pivot.head()
```

```python
# 1. Calculate Energy Security Percentage
# (Generation / Demand) * 100
df_pivot['energy_security_pct'] = (df_pivot['generation_twh'] / df_pivot['demand_twh']) * 100

# 2. Basic Cleaning: Remove rows with NaN (where one of the two values was missing)
df_pivot = df_pivot.dropna(subset=['energy_security_pct'])

# 3. Quick EDA: Look at the most "secure" countries in the most recent year
latest_year = df_pivot['Year'].max()
print(f"--- Top 5 Energy Secure Countries in {latest_year} ---")
display(df_pivot[df_pivot['Year'] == latest_year].sort_values('energy_security_pct', ascending=False).head())
```

```python
# 1. CLEANING: Filter for the correct parameter and mode
# Use 'EV sales' (lowercase) as verified in your previous check
plot_df = ev_data[ev_data['parameter'] == 'EV sales'].copy()

# 2. GROUPING: Map countries to the groups in your target legend
# To avoid double-counting, we label 'World' as its own segment
def map_region(r):
    if r == 'China': return 'China'
    if r in ['USA', 'Canada', 'Mexico']: return 'North America'
    if r == 'World': return 'World'
    return 'Other'

plot_df['Grouped_Region'] = plot_df['region'].apply(map_region)

# 3. PREDICTION: Create the label for actual vs. predicted years
# Categories found: 'Historical', 'Projection-STEPS', 'Projection-APS'
plot_df['Prediction'] = plot_df['category'].apply(
    lambda x: 'Actual' if x == 'Historical' else 'Predicted'
)

# 4. FIX AXIS: Convert year to string to prevent the -1, 0, 1 scale
# Also filter out the 'World' region if you want to see the regional breakdown
# OR keep it if you want to see it as a base layer like your screenshot
plot_df['year'] = plot_df['year'].astype(str)

# 5. AGGREGATE: Sum values to ensure one bar per year/region group
final_df = plot_df.groupby(['year', 'Grouped_Region', 'Prediction'], as_index=False)['value'].sum()

# 6. PLOT: Match the 'ggplot2' style and legend
fig = px.bar(final_df, 
             x='year', 
             y='value', 
             color='Grouped_Region',
             pattern_shape='Prediction', # Adds texture to future years
             title='EV Sales Analysis: Historical vs. Predicted',
             labels={'value': 'Number of EV Sales', 'year': 'Year'},
             template='ggplot2',
             # Ensure years are in order: 2010...2023, 2025, 2030, 2035
             category_orders={"year": sorted(final_df['year'].unique())})

# Force the axis to treat years as discrete labels
fig.update_xaxes(type='category', tickangle=45)
fig.update_layout(barmode='stack')

fig.show()
```

```python

def get_ev_visual(df, parameter):
    """
    Input a dataframe and a parameter name to get a regional trend graph.
    """
    # 1. Cleaning & Initial Filter
    # Filter for the specific parameter passed in the function call
    # We exclude 'World' to avoid double-counting the stack
    mask = (df['parameter'] == parameter) & (df['region'] != 'World')
    plot_df = df[mask].copy()

    # 2. Region Mapping
    region_map = {
        'China': 'China',
        'USA': 'North America',
        'Europe': 'Europe',
        'India': 'Other',
        'Australia': 'Other'
    }
    plot_df['Grouped_Region'] = plot_df['region'].map(region_map).fillna('Other')

    # 3. Prediction Status
    # Logic to separate 'Actual' from 'Predicted' for future years
    plot_df['Prediction'] = plot_df['category'].apply(
        lambda x: 'Actual' if x == 'Historical' else 'Predicted'
    )

    # 4. Fix Year Axis (String conversion)
    plot_df['year'] = plot_df['year'].astype(str)
    
    # Aggregate data to ensure one clean bar per year/region group
    final_df = plot_df.groupby(['year', 'Grouped_Region', 'Prediction'], as_index=False)['value'].sum()

    # 5. Visualization
    fig = px.bar(final_df, 
                 x='year', 
                 y='value', 
                 color='Grouped_Region',
                 pattern_shape='Prediction',
                 title=f'Trends: {parameter}',
                 labels={'value': parameter, 'year': 'Year'},
                 template='ggplot2',
                 category_orders={"year": sorted(final_df['year'].unique())})

    fig.update_xaxes(type='category', tickangle=45)
    fig.update_layout(barmode='stack')
    
    return fig.show()

# --- HOW TO CALL IT ---
# Example 1: Using 'EV sales'
get_ev_visual(ev_data, parameter='EV sales')

# Example 2: Using 'EV stock'
get_ev_visual(ev_data, parameter='EV stock')
```

```python
def explain_ev_trends(df, parameter='EV sales'):
    """
    Generates a summary table to explain the regional trends.
    """
    # 1. Reuse your cleaning logic
    plot_df = df[(df['parameter'] == parameter) & (df['region'] != 'World')].copy()
    
    region_map = {'China': 'China', 'USA': 'North America', 'Europe': 'Europe'}
    plot_df['Grouped_Region'] = plot_df['region'].map(region_map).fillna('Other')
    
    # 2. Pivot the data to get a Year-over-Year view
    summary = plot_df.groupby(['year', 'Grouped_Region'])['value'].sum().unstack()
    
    # 3. Calculate Totals and Growth Rates
    summary['Global_Total'] = summary.sum(axis=1)
    summary['YoY_Growth_%'] = summary['Global_Total'].pct_change() * 100
    
    # 4. Filter for key years to keep it readable (e.g., 2023 vs 2035)
    key_years = [2010, 2020, 2023, 2030, 2035]
    available_keys = [y for y in key_years if y in summary.index]
    
    return summary.loc[available_keys].round(2)

# Call it to get your "talking points"
stats = explain_ev_trends(ev_data)
print(stats)
```

```python
def get_accurate_market_share(df, region='China', year=2023):
    # 1. Filter for specific mode and parameter to avoid summing percentages
    # We use 'Cars' to match your Energy Dashboard focus
    mask = (
        (df['region'] == region) & 
        (df['year'] == year) & 
        (df['parameter'] == 'EV sales share') & 
        (df['mode'] == 'Cars') & 
        (df['powertrain'] == 'EV') # 'EV' usually represents the total of BEV + PHEV
    )
    
    result = df[mask]
    
    if not result.empty:
        share = result['value'].values[0]
        # Check if the CSV uses 0.38 or 38
        real_percent = share if share > 1 else share * 100
        print(f"Correct Market Share for {region} ({year}): {real_percent:.1f}%")
    else:
        print("No matching data found. Check your filter strings!")

get_accurate_market_share(ev_data)
```

```python
import pandas as pd
import plotly.express as px

def plot_china_actual_ev(df, Region):
    # 1. Filter for China, Cars, and ACTUAL historical sales
    actual_china = df[
        (df['region'] == Region) & 
        (df['mode'] == 'Cars') & 
        (df['parameter'] == 'EV sales') & 
        (df['category'] == 'Historical') # This removes the predicted data
    ].copy()

    # 2. Exclude the total 'EV' powertrain to see the BEV vs PHEV split
    actual_china = actual_china[actual_china['powertrain'] != 'EV']

    # 3. Format years
    actual_china['year'] = actual_china['year'].astype(str)
    actual_china = actual_china.sort_values('year')

    # 4. Create the Plot
    fig = px.bar(actual_china, 
                 x='year', 
                 y='value', 
                 color='powertrain',
                 title='China: Actual Historical EV Sales (BEV vs PHEV)',
                 labels={'value': 'Number of Vehicles', 'year': 'Year'},
                 template='ggplot2',
                 color_discrete_map={'BEV': '#1f77b4', 'PHEV': '#ff7f0e'})

    fig.update_xaxes(type='category')
    fig.update_layout(barmode='stack')
    
    return fig.show()

# Run the function
plot_china_actual_ev(ev_data, 'China')
```
