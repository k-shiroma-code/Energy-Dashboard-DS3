```python
# 1. Define the variables we need for the Energy Security task
security_vars = ['Demand', 'Total Generation']

# 2. Filter the Global and Europe DataFrames
# This makes the data much smaller and faster to work with
df_global_security = df_global_yearly[df_global_yearly['Variable'].isin(security_vars)]
df_europe_security = df_europe_yearly[df_europe_yearly['Variable'].isin(security_vars)]

# 3. Quick check to see the first few rows
print(f"Global Security Data Rows: {len(df_global_security)}")
display(df_global_security.head())
```

```python
# Create the pivoted 'Wide' DataFrame
df_pivot = df_global_security.pivot_table(
    index=['Area', 'ISO 3 code', 'Year', 'Continent'], 
    columns='Variable', 
    values='Value'
).reset_index()

# Rename columns for easier math
df_pivot.columns.name = None  # Remove the hierarchy name
df_pivot = df_pivot.rename(columns={
    'Demand': 'demand_twh',
    'Total Generation': 'generation_twh'
})

# Preview the transformation
df_pivot.head()
```

```python
# 1. Calculate Energy Security Percentage
# (Generation / Demand) * 100
df_pivot['energy_security_pct'] = (df_pivot['generation_twh'] / df_pivot['demand_twh']) * 100

# 2. Basic Cleaning: Remove rows with NaN (where one of the two values was missing)
df_pivot = df_pivot.dropna(subset=['energy_security_pct'])

# 3. Quick EDA: Look at the most "secure" countries in the most recent year
latest_year = df_pivot['Year'].max()
print(f"--- Top 5 Energy Secure Countries in {latest_year} ---")
display(df_pivot[df_pivot['Year'] == latest_year].sort_values('energy_security_pct', ascending=False).head())
```

```python
# 1. Get current Renewable Capacity (GW) from the main yearly dataset
# Assuming 'Capacity' is a Category or Subcategory in your Ember file
df_current_res = df_global_yearly[
    (df_global_yearly['Year'] == 2024) & 
    (df_global_yearly['Variable'] == 'Renewables') &
    (df_global_yearly['Unit'] == 'GW')
].copy()

# Rename for clear merging
df_current_res = df_current_res[['ISO 3 code', 'Value']].rename(columns={'Value': 'current_capacity_gw'})
```

```python
# 1. Filter for the two variables needed for the Energy Security task
security_vars = ['Demand', 'Total Generation']
df_filtered = df_global_yearly[df_global_yearly['Variable'].isin(security_vars)]

# 2. Pivot the data to get Demand and Generation in side-by-side columns
df_pivot = df_filtered.pivot_table(
    index=['Area', 'ISO 3 code', 'Year', 'Continent'], 
    columns='Variable', 
    values='Value'
).reset_index()

# 3. Clean column names and calculate the Energy Security %
df_pivot.columns.name = None
df_pivot = df_pivot.rename(columns={
    'Demand': 'demand_twh',
    'Total Generation': 'generation_twh'
})

# Energy Security formula: (Generation / Demand) * 100
df_pivot['energy_security_pct'] = (df_pivot['generation_twh'] / df_pivot['demand_twh']) * 100
```

```python
# 1. Inspect actual columns to find the correct 'country_code' equivalent
print("Current Target Columns:", target_2030.columns.tolist())

# 2. Rename specifically for the merge. 
# Based on your previous preview, use the exact names found in your file:
target_2030 = target_2030.rename(columns={
    'country_code': 'ISO 3 code', # Use the exact key from your df_pivot
    'country_name': 'Area',
    'Renewables': 'target_2030_gw'
})
```

```python
# 1. Strip whitespace from all column names to remove hidden spaces
target_2030.columns = target_2030.columns.str.strip()

# 2. Select the cleaned columns
# Now 'Renewables' will be recognized correctly
df_targets_subset = target_2030[['ISO 3 code', 'Renewables']].copy()

# 3. Rename for your dashboard's "Source of Truth"
df_targets_subset = df_targets_subset.rename(columns={'Renewables': 'target_2030_gw'})

# 4. Merge onto your calculated security dataframe
# This links your 2000-2024 performance with 2030 goals
df_final = pd.merge(df_pivot, df_targets_subset, on='ISO 3 code', how='left')

# 5. Final cleaning for the dashboard
df_final = df_final.replace([np.inf, -np.inf], np.nan).dropna(subset=['energy_security_pct'])

print("‚úÖ Data successfully integrated into df_final!")
display(df_final.head())
```
